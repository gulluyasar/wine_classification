```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors=20)
outlier_scores = lof.fit_predict(df)

outlier_indices = np.where(outlier_scores == -1)[0]

print("Aykırı Değer Sayısı:", len(outlier_indices))
print("Aykırı Değerlerin İndeksleri:", outlier_indices)
```

```python
df.loc[outlier_indices]
```

```python
len(outlier_indices) / df.shape[0] * 100
```

```python
#silmek 
# no_liar_df = df.drop(outlier_indices, axis=0)
# doldurmak 
mean_values = df.mean()

df_filled = df.copy()  
for col in df.columns:
    df_filled.iloc[outlier_indices, df.columns.get_loc(col)] = mean_values[col]


print("Mean values used for filling outliers:")
print(mean_values)


print("\nDataframe with outliers filled:")
print(df_filled.head()) 
```

```python
# no_liar_df['quality'].value_counts()
df_filled['quality'].value_counts()
```

```python
# df = no_liar_df.copy()
df = df_filled.copy()
df['quality'] = pd.cut(df['quality'], bins=3, labels=False) #???
```

```python
# veri özellik çıkarımı 
from sklearn.feature_selection import SelectKBest,f_regression
X = df.drop("quality",axis=1)
y = df['quality']

selector = SelectKBest(k=5,score_func=f_regression)
selector.fit(X,y)
df.columns[selector.get_support(indices=True)]
```

```python
X = df_[list(df.columns[selector.get_support(indices=True)])]
y = df_['quality']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

models = {'RF': RandomForestClassifier(),
         'GB': GradientBoostingClassifier(),
         'SVM': SVC(),
         'XGB': xgb.XGBClassifier()}
```

```python
for modelName, model in models.items():
    model.fit(X_train, y_train)
    acc = accuracy_score(y_test, model.predict(X_test))
    print(f'{modelName}: {acc}')
```

```python
svm = SVC()
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
```

```python
grid = GridSearchCV(svm, cv=5, param_grid=param_grid, refit=True, verbose=3)
grid.fit(X_train, y_train)
```

```python
grid.score(X_test, y_test)
```

```python
grid.best_params_
```

```python
model = grid.best_estimator_
model
```

```python
model.score(X_test, y_test)
```

```python
son_model = SVC(C=100, gamma=0.1)
son_model.fit(X_train, y_train)
predict = son_model.predict(X_test)
accuracy_score(y_test, son_model.predict(X_test))
```
